```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
ratings = pd.read_csv('u.data',sep='\t',names=
                     ['user_id','item_id','rating','timestamp'])
```

```python
ratings.head()
```

```python
movies=pd.read_csv('u.item',sep='|',encoding='latin-1',header=None)
```

```python
movies= movies.iloc[:,[0,1]]
movies.head()
```

```python
movies.columns=['movie_id','movie_title']
```

```python
movies.head(10)
```

```python
ratings.head(10)
```

```python
ratings.drop('timestamp',axis=1,inplace=True)
```

# Building Association Matrix

```python
learningMatrix= ratings.pivot_table(index=['item_id'],columns=['user_id'],values=['rating']).reset_index(drop=True)
```

```python
learningMatrix.fillna(0,inplace=True)
```

```python
learningMatrix
```

# Finding Similarity between the movies(items)

```python
from sklearn.metrics import pairwise_distances
```

```python
movie_similarity = pairwise_distances(learningMatrix,metric='cosine')
np.fill_diagonal(movie_similarity,0)
```

```python
similarity_matrix = pd.DataFrame(movie_similarity)
similarity_matrix.head(15)
```

# Recommendation System in Action

```python
movies[movies['movie_title']=='Copycat (1995)'].index.tolist()
```

```python
try:
    user_inp=input('Enter the reference movie title based on which recommendations are to be made : ')
    
    inp=movies[movies['movie_title']==user_inp].index.tolist()
    
    inp=inp[0]
    
    movies['similarity']=similarity_matrix.iloc[inp]
    
    print('Recommended movies based on your choice of',user_inp,': \n',movies.sort_values(['similarity'],ascending=False)[1:10])

except:
    print('The movie name you entered does not exist in the list, however, below are the top movies recommended in general')
    print(movies.head(10))
```
